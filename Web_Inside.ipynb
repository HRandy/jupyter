```python
import requests as r
from bs4 import BeautifulSoup
url='http://www.inside.com.tw/page/{}'
for page in range(1,30):
    inside=r.get(url.format(page))
    soup=BeautifulSoup(inside.text,'lxml')
    articles=soup.select('.post_list_item_content')
    
    
    for article in articles:
        article_dict={}
        
        article_dict['title']=article.select_one('.js-auto_break_title').text
        article_dict['contect']=article.select_one('.post_description').text
        print '-------------------------'
        
        for k in article_dict:
            print article_dict[k]
        
       
```
