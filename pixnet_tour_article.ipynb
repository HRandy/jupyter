```python
import requests as r
from bs4 import BeautifulSoup
WEBSITE_URL = 'https://www.pixnet.net/blog/articles/category/28/new/{}'
dict={}
url_list=[]
for pages in range(1,1000):
    res = r.get(WEBSITE_URL.format(pages))
    res.encoding='utf-8'
    soup = BeautifulSoup(res.text, 'lxml')

    article_block = soup.select('.article-list')[0].select('li')    #big block


    for i in article_block:                                     #url
        dict['url']=i.select_one('a')['href']
        url_list.append(i.select_one('a')['href'])
    #     a.append(dict)
        for j in url_list:
            
            a=r.get(j)
            a.encoding='utf-8'
            soup = BeautifulSoup(a.text, 'lxml')
            title = soup.select('.article')[0].select('ul')[0].text.strip()

            content = soup.select('.article')[0].select('.article-content-inner')[0].text.strip()
            print title,content
```
