```python
## 抓取網頁起手式
import requests as r
import json
from bs4 import BeautifulSoup
res = r.get("https://www.ptt.cc/bbs/Japan_Travel/")
soup = BeautifulSoup(res.text, 'lxml')

##  抓出最新一頁的頁數
ptt_list =soup.select('div.btn-group-paging')[0].select('a')
a = ptt_list[1]['href']
page = a.split('index')[1].split('.html')[0]

articles= []
## 抓出最後1頁的相關內容
for i in range (100,150):   #################################################################  改數字就可以多抓幾頁

    URL="https://www.ptt.cc/bbs/Japan_Travel/index{}.html".format(i)
    
    res2 = r.get(URL)
    soup2 = BeautifulSoup(res2.text, 'lxml')
    ptt2_list=soup2.select('div.r-ent')

    for article in ptt2_list:
        article_dict = {}
        try:
            article_dict['type']  = article.select_one('div.title').text.split(']')[0].split('[')[1].strip()
            article_dict['title'] = article.select_one('div.title').text.strip()
            webURL="https://www.ptt.cc"+article.select_one('div.title').select('a')[0]['href'].strip()
            article_dict['URL']   = webURL
            res3 = r.get(webURL)
            soup2 = BeautifulSoup(res3.text, 'lxml')
            article_dict['date']= soup2.select('.article-meta-value')[3].text
            article_dict['Author']= article.select_one('div.author').text.strip()

        except Exception as e:
             print e
        articles.append(article_dict)

## 存成Json 檔
pttjson = json.dumps(articles, ensure_ascii=False) # This is a unicode string
with open('ptt100150.json', 'w') as f: ###################################################### 爬到哪一篇就存成檔名
    f.write(pttjson.encode('utf-8'))
```
